## Importing the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import *
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
import warnings
warnings.filterwarnings('ignore')
from pathlib import Path
import joblib
```

## Data Collection and Preprocessing

```python
# Load the dataset using Pandas Dataframe
data_path=Path("../data/diabetes.csv")
d=pd.read_csv(data_path)
```

0-> No Diabetes

```python
cols = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']

for c in cols:
    d[c] = d[c].replace(0, d[c].median())
```

## Train Test Split

```python
# Creating the independent and dependent variable
X=d.drop(columns=['Outcome'])
y=d['Outcome']
```

```python
# Splitting into test(20%) and train(80%) Data
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

## Logistic Regression

```python
# Training with the logistic Model
lm=LogisticRegression(class_weight='balanced')
lm.fit(X_train,y_train)
```

```python
# Predicting the testing data
lm_pred=lm.predict(X_test)
```

```python
# Getting the Accuracy Score
accuracy_score(y_test,lm_pred)
```

```python
# Printing the Classification Report
print(classification_report(y_test, lm_pred))
```

```python
# Printing the Confusion Matrix
confusion_matrix(y_test, lm_pred)
```

## Random Forest

```python
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
rf_pred=rf.predict(X_test)
```

```python
accuracy_score(y_test,rf_pred)
```

```python
print(classification_report(y_test, rf_pred))
```

## SVM

```python
sm=SVC()
sm.fit(X_train,y_train)
```

```python
sm_pred=sm.predict(X_test)
```

```python
accuracy_score(y_test,sm_pred)
```

```python
print(classification_report(y_test, sm_pred))
```

## XGBoost

```python
xm=XGBClassifier(n_estimators=200,
    learning_rate=0.05,
    max_depth=4,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42)
xm.fit(X_train,y_train)
```

```python
xm_pred=xm.predict(X_test)
```

```python
accuracy_score(y_test,xm_pred)
```

```python
print(classification_report(y_test, xm_pred))
```

## Model Selection

SVM was selected as the final model because it achieved a higher recall for the diabetic class, reducing false negatives and ensuring better detection of potential diabetes cases. In healthcare applications, recall is prioritized to avoid missing patients who may have the disease.

```python
# Save the trained Model
joblib.dump(sm,"diabetes_model.pkl")
```
